```
from google.cloud import bigquery
```

```
from google.colab import auth
auth.authenticate_user()
print('Authenticated')
```

```
#Creating a client object
project_id = "sqlproject-302612"
client = bigquery.Client(project = project_id)
```

```
dataset_ref = client.dataset("nhtsa_traffic_fatalities", project="bigquery-public-data")
dataset = client.get_dataset(dataset_ref)
```

```
tables = client.list_tables(dataset)
for table in tables:
  print(table.table_id)
```

```
# fetching accident_2015 table
table_ref = dataset_ref.table("accident_2015")
table = client.get_table(table_ref)
```

```
# visualizing first rows
client.list_rows(table, max_results=5).to_dataframe()
```

```
# Constructing the query
query = ''' SELECT EXTRACT(DAYOFWEEK FROM timestamp_of_crash) AS day_of_week,
                   COUNT(consecutive_number) AS num_accidents
            FROM `bigquery-public-data.nhtsa_traffic_fatalities.accident_2015`
            GROUP BY day_of_week
            ORDER BY num_accidents DESC
        '''
```

```
#Submmiting the query
query_job = client.query(query)
day_most_fatality = query_job.to_dataframe()
day_most_fatality.head()
```
